import pandas as pd

df = pd.DataFrame({
    'Age': [25, 32, 47, 51, 62],
    'Salary': [40000, 50000, 60000, 52000, 62000],
    'Purchased': [0, 1, 1, 0, 1]
})

X = df[['Age', 'Salary']]
y = df['Purchased']
Use Mutual Information to find the most important feature for predicting Purchased. Select the top 1 feature.

Expected Output:
['Salary']

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, chi2
df = pd.DataFrame({
    'Age': [25, 32, 47, 51, 62],
    'Salary': [40000, 50000, 60000, 52000, 62000],
    'Purchased': [0, 1, 1, 0, 1]
})
X = df[['Age', 'Salary']]
y = df['Purchased']
features=SelectKBest(score_func=chi2,k=1)
features.fit_transform(X,y)
selected_features=X.columns[features.get_support()]
print(selected_features.to_list())
```

import pandas as pd

data = pd.DataFrame({
    'User': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Item': ['Apple', 'Banana', 'Apple', 'Orange', 'Banana', 'Orange'],
    'Rating': [5, 3, 4, 2, 5, 4]
})

Based on item popularity (sum of ratings), recommend top 2 items to a new user.

 Expected Output:

['Apple', 'Banana']



```python
import pandas as pd

data = pd.DataFrame({
    'User': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Item': ['Apple', 'Banana', 'Apple', 'Orange', 'Banana', 'Orange'],
    'Rating': [5, 3, 4, 2, 5, 4]
})
recommended_item=data.groupby('Item')['Rating'].sum().sort_values(ascending=False).head(2).index.to_list()
print(recommended_item)
```

from sklearn.datasets import load_diabetes
import pandas as pd

X, y = load_diabetes(return_X_y=True, as_frame=True)
Write a Python program to perform feature selection using Recursive Feature Elimination (RFE). Use Linear Regression as the estimator and select the top 5 features. Print the selected feature names.

Expected output:
Selected features: ['bmi', 's5', 'bp', 'age', 'sex']


```python
from sklearn.datasets import load_diabetes
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
X, y = load_diabetes(return_X_y=True, as_frame=True)
feature=RFE(estimator=LinearRegression(),n_features_to_select=5)
feature.fit_transform(X,y)
selected_feature=X.columns[feature.get_support()].to_list()
print("Selected features:",selected_feature)
```

from sklearn.datasets import load_digits
import pandas as pd

X, y = load_digits(return_X_y=True, as_frame=True)
Write a Python program that selects the top 8 features using SelectKBest with mutual_info_classif as the scoring function. Print the names of the selected features.

ðŸŽ¯ Expected Output:
    Top 8 features using mutual info: ['pixel_0_1', 'pixel_0_6', 'pixel_1_5', 'pixel_2_2', 'pixel_2_5', 'pixel_4_2', 'pixel_5_1', 'pixel_7_4']


```python
from sklearn.datasets import load_digits
import pandas as pd
from sklearn.feature_selection import SelectKBest,mutual_info_classif
X, y = load_digits(return_X_y=True, as_frame=True)
feature=SelectKBest(score_func=mutual_info_classif,k=8)
feature.fit_transform(X,y)
selected_features=X.columns[feature.get_support()].to_list()
print("Top 8 features using mutual info:",selected_features)
```

from sklearn.datasets import load_breast_cancer
import pandas as pd

X, y = load_breast_cancer(return_X_y=True, as_frame=True)
Write a Python program to perform feature selection using SelectFromModel. Train a Logistic Regression model and select features based on importance using SelectFromModel. Print the names of the selected feature


Expected Output:
Selected features: ['mean radius', 'mean concavity', 'texture error', 'worst texture', 'worst smoothness', 'worst compactness', 'worst concavity', 'worst concave points', 'worst symmetry']

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel

X, y = load_breast_cancer(return_X_y=True, as_frame=True)
feature=SelectFromModel(estimator=LogisticRegression(max_iter=10000),threshold="mean")
feature.fit_transform(X,y)
selected_feature=X.columns[feature.get_support()].to_list()
print("Selected features:",selected_feature)
```
